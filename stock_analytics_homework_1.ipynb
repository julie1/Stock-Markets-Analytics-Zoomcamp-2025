```python
import re
import pandas as pd

def extract_from_parentheses(text):
  """
  Extracts the content within parentheses from a string.

  Args:
    text: The input string.

  Returns:
    The content within parentheses, or None if no parentheses are found.
  """
  match = re.search(r"\((.*?)\)", text)
  if match:
    return match.group(1)
  return None
```

```python
def pairwise(iterable):
    """Return successive overlapping pairs from an iterable."""
    a = iter(iterable)
    try:
        first = next(a)
        while True:
            second = next(a)
            yield (first, second)
            first = second

    except StopIteration:
        return
    
```

```python
def triplewise(iterable):
    a = iter(iterable)
    try:
        first = next(a)
        second = next(a)
        while True:
            third = next(a)
            yield first, second, third
            first, second = second, third
    except StopIteration:
        return

```

```python
ad = amzn_data.copy()
ad.columns = [amzn_data.columns[i][0] for i in range(len(amzn_data.columns))]
ad.rename(columns={'Adj Close': 'Adj_Close'}, inplace=True)
print(ad.columns)
ad.reset_index(inplace=True)
print(ad.head())
#print(hd['Date'])


ad = ad.astype({'Date': "string", 'Adj_Close': float, 'Close': float, 'High': float, 'Low': float, 'Open': float, 'Volume': int})
print(ad.dtypes)
```

```python
ad1 = []
for row in ad.itertuples():
   # print(row, type(row))
    rowc = row.Adj_Close
    rowd = row.Date
    ad1.append((rowd, rowc))
  
```

```python
returns = pd.DataFrame(columns=['date', 'twoday_return'])
returns = returns.astype({'date': "string", 'twoday_return': float})
for date1, date2, date3 in triplewise(ad1):
    p1 = date1[1]
    p3 = date3[1]
    date0 = date2[0]
    twoday_return = p3/p1 - 1
   # print(date0, twoday_return)
    returns = pd.concat([returns, pd.DataFrame({'date': [date0], 'twoday_return': [twoday_return]})])
```

```python
def nearest(items, pivot):
    return min(items, key=lambda x: abs(x - pivot))

bullorbear = pd.DataFrame(columns=['date', 'market'])
for row in surprise_returns.itertuples():
    day = dt.strptime(row.date, "%Y-%m-%d")
    closest = nearest(mindates, day)
    if closest >= day: bullorbear = pd.concat([bullorbear, pd.DataFrame({'date': [row.date], 'market': ['bear']})])
    else: bullorbear = pd.concat([bullorbear, pd.DataFrame({'date': [row.date], 'market': ['bull']})])
                                              
    
```

```python
surprises = pd.merge(surprise_returns, bullorbear, on='date')
#print(surprises)
bears = surprises[surprises['market']=='bear']
bulls = surprises[surprises['market']=='bull']
median_bear_return = bears['twoday_return'].quantile(0.5)
median_bull_return = bulls['twoday_return'].quantile(0.5)
print('The median bear return is: ', "{:.2%}".format(median_bear_return))
print('The median bull return is: ', "{:.2%}".format(median_bull_return))
```
